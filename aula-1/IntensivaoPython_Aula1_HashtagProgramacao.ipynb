### Passo 1 - Importando a Base de Dados

```
import pandas as pd
df_sales = pd.read_excel(r'/content/drive/MyDrive/Colab Notebooks/IntensivaÃÉo de Python/Aula 1/Sales.xlsx')
```

### Passo 2 - Visualizando os Dados

```
display(df_sales)
```

### Passo 3.1 - Calculando o Faturamento por Loja

```
billing = df_sales[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
billing = billing.sort_values(by='Valor Final', ascending=False)
display(billing)
```

### Passo 3.2 - Calculando a Quantidade Vendida por Loja

```
quantity = df_sales[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()
quantity = quantity.sort_values(by='ID Loja', ascending=False)
display(quantity)
```

### Passo 3.3 - Calculando o Ticket M√©dio dos Produtos por Loja

```
ticket_average = (billing['Valor Final'] / quantity['Quantidade']).to_frame()
ticket_average = ticket_average.rename(columns={0: 'Ticket Medio'})
ticket_average = ticket_average.sort_values(by='Ticket Medio', ascending=False)
display(ticket_average)
```

### Criando a Fun√ß√£o de Enviar e-mail

```
import smtplib
import email.message

def send_email(shop_summary, shop):
  server = smtplib.SMTP('smtp.gmail.com:587')  
  email_content = f'''
  <p>Prezados,</p>
  {shop_summary.to_html()}
  <p>At.te,</p>
  <p>Bryan Lima</p>
  <p>Software Developer</p>
  '''
  
  msg = email.message.Message()
  msg['Subject'] = f'Relat√≥rio - {shop}'
    
  msg['From'] = SOURCE_EMAIL
  msg['To'] = DESTINATION_EMAIL

  msg.add_header('Content-Type', 'text/html')
  msg.set_payload(email_content)
  
  s = smtplib.SMTP('smtp.gmail.com: 587')
  s.starttls()
  
  s.login(msg['From'], PASSWORD_EMAIL)
  s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
```

### Passos 4 e 5 - Calculando Indicadores por Loja + Enviando E-mail para Cada Loja

```
shops = df_sales['ID Loja'].unique()

for shop in shops:
  table_shop = df_sales.loc[df_sales['ID Loja'] == shop, ['ID Loja', 'Quantidade', 'Valor Final']]
  shop_summary = table_shop.groupby('ID Loja').sum()
  shop_summary['Ticket Medio'] = shop_summary['Valor Final'] / shop_summary['Quantidade']
  send_email(shop_summary, shop)
```

### Passo 6 - Enviando E-mail para Diretoria

```
board_table = billing.join(quantity).join(ticket_average)
send_email(board_table, 'Todas as Lojas')
```
